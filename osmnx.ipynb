```python
import osmnx as ox
```

```python
# Получение графа дорожной сети Хасана (Россия)
place_name = "Кавалерово, Россия"
#place_name = "Петропавловка, Россия"
G = ox.graph_from_place(place_name, network_type="all", simplify=True, retain_all=True)
```

```python
#Визуализация графа с белым фоном, синими рёбрами и черными вершинами
ox.plot_graph(G, bgcolor="white", node_color="black", edge_color="blue", node_size=10, show=True)
```

```python
import matplotlib.pyplot as plt
```

```python
# Координаты целевой точки (долгота, широта)
X = 130.640518  # Пример долготы
Y = 42.430995    # Пример широты


# Поиск ближайшей вершины
nearest_node = ox.distance.nearest_nodes(G, X, Y)
print(f"Ближайшая вершина: {nearest_node}")
```

```python
# Визуализация графа с выделением ближайшей вершины
# ox.plot_graph возвращает два объекта:
#   fig — объект Figure, представляющий весь холст для графика
#   ax  — объект Axes, на котором непосредственно рисуется граф
fig, ax = ox.plot_graph(
    G,
    bgcolor="white",    # Цвет фона графика
    node_color="blue",  # Цвет всех вершин
    node_size=10,       # Размер всех вершин
    edge_color="gray",  # Цвет рёбер
    show=False,         # Не показывать график сразу (для дальнейшей настройки)
    close=False,        # Не закрывать график после отрисовки (чтобы можно было добавить элементы)
)

# Получаем координаты ближайшей вершины из графа по её идентификатору
nearest_node_x = G.nodes[nearest_node]["x"]  # Долгота вершины
nearest_node_y = G.nodes[nearest_node]["y"]  # Широта вершины

# Добавляем на оси (ax) красную точку для выделения ближайшей вершины
ax.scatter(
    nearest_node_x,
    nearest_node_y,
    c="red",           # Цвет точки (красный)
    s=100,             # Размер точки
    alpha=0.8,         # Прозрачность точки
    edgecolor="black", # Цвет границы точки
    linewidth=1,       # Толщина границы точки
    zorder=3,          # Порядок отрисовки, чтобы точка была поверх остальных элементов графа
)

# Показываем окончательный график
plt.show()
```

```python
import networkx as nx
```

```python
G = ox.graph_from_place(place_name, network_type="drive", simplify=True, retain_all=False)
ox.plot_graph(G, bgcolor="white", node_color="black", edge_color="blue", node_size=10, show=True)
# Вывод базовой информации о графе
num_vertices = len(G.nodes)
num_edges = len(G.edges)
print(f"Число вершин (V): {num_vertices}")
print(f"Число рёбер (E): {num_edges}")
```

```python
#Преобразуйте в матрицу смежности
adjacency_matrix = nx.to_numpy_array(G, weight='length')
print(adjacency_matrix)
```

```python
adjacency_df = nx.to_pandas_adjacency(G, weight='length', dtype=float)
print(adjacency_df)
```

```python
# Выполнение алгоритма Дейкстры от вершины nearest_node:
# Возвращается кортеж из двух словарей:
# 1. lengths: кратчайшее расстояние от nearest_node до каждой вершины
# 2. paths: список вершин (маршрут) от nearest_node до каждой вершины
lengths, paths = nx.single_source_dijkstra(G, source=nearest_node, weight='length')
print(lengths)
print(paths)
```

```python
import random
some_nodes = random.sample(list(G.nodes), 5)
print(f"Кратчайшие пути от вершины {nearest_node}:")
for n in some_nodes:
    if n in paths:
        print(f"Вершина {n}: расстояние {lengths[n]}, путь: {paths[n]}")
```

```python
# номер (идентификатор) вершины, координаты которой нужно вывести
node_id = 1226395462

# Проверяем, что вершина существует в графе G
if node_id in G.nodes:
    node_data = G.nodes[node_id]
    longitude = node_data['x']
    latitude = node_data['y']
    print(f"Координаты вершины {node_id}: долгота = {longitude}, широта = {latitude}")
else:
    print(f"Вершина с номером {node_id} не найдена в графе.")
```

```python
# идентификаторы начальной и конечной вершин
start_node = 1226395594  # пример идентификатора начальной вершины
end_node   = 2584089272  # пример идентификатора конечной вершины

# Проверяем наличие вершин в графе
if start_node not in G.nodes:
    print(f"Начальная вершина {start_node} не найдена в графе.")
elif end_node not in G.nodes:
    print(f"Конечная вершина {end_node} не найдена в графе.")
else:
    # Выполнение алгоритма Дейкстры с использованием single_source_dijkstra
    # Если указан параметр target, функция возвращает (расстояние, путь)
    distance, path = nx.single_source_dijkstra(G, source=start_node, target=end_node, weight='length')
    print("Найденный маршрут:", path)
    print("Общее расстояние:", distance)

    # Визуализация найденного маршрута
    fig, ax = ox.plot_graph_route(G, path, bgcolor="white", node_color="black", edge_color="blue", route_linewidth=3, node_size=10)
    plt.show()
```
